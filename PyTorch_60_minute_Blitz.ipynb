# Tensors

Tensors are a specialized data structure that are very similar to arrays and matrices. In PyTorch, we use tensors to encode the input and outputs of a model, as well as the model's parameters.
<br><br>

Tensors are similar to NumPy's ndarrays, except that tensors can run on GPU's or other specialized hardware to accelerate computing.

```python
# Import PyTorch and NumPy
import torch
import numpy as np
```

```python
torch.__version__
```

```python
np.__version__
```

## Tensor Initialization

Tensors can be initialized in various ways

### **Directly from data**

Tensors can be created directly from data. The data type is automatically infered.

```python
data = [[1, 2], [3, 4]]
x_data = torch.tensor(data)
x_data
```

```python
# Data type of tensor
x_data.dtype
```

```python
# Shape of tensor
x_data.shape
```

### **From a NumPy array**

Tensors can be created from NumPy arrays (and vice versa - see [Bridge with NumPy](https://docs.pytorch.org/tutorials/beginner/blitz/tensor_tutorial.html#bridge-to-np-label))

```python
np_array = np.array(data)
print(np_array)
x_np = torch.from_numpy(np_array)
x_np
```

### **From another tensor**

The new tensor retains the properties (`shape`, `datatype`) of the argument tensor, unless explicitly orverridden

```python
# Create ones tensor simillar to x_data
x_ones = torch.ones_like(x_data) # Retains the properties of x_data
print(f"Ones Tensor: \n {x_ones} \n")

# Create zero tensor simillar to x_data
x_zeros = torch.zeros_like(x_data)
print(f"Zeros Tensor: \n {x_zeros} \n")

# Create random tensor simillar to x_data
x_rand = torch.rand_like(x_data, dtype = torch.float) # Changing the data type to float
print(f"Random Tensor: \n {x_rand} \n")
```

### **With random or constant values**

`shape` is a tuple of tensor dimensions. In the functions below, it determines the dimensionality of the output tensor

```python
shape = (2, 3)
rand_tensor = torch.rand(shape)
ones_tensor = torch.ones(shape)
zeros_tensor = torch.zeros(shape)

print(f"Random Tensor: \n {rand_tensor} \n")
print(f"Ones Tensor: \n {ones_tensor} \n")
print(f"Zeros Tensor: n {zeros_tensor} \n")
```

## Tensor Attributes

Tensor attributes describes their shape, datatype and the device on which they are stored

```python
tensor = torch.rand(3, 4)
print(f"Tensor: \n {tensor} \n")

# Attributes
print(f"Shape of tensor: {tensor.shape}")
print(f"Datatype of tensor: {tensor.dtype}")
print(f"Device tensor is stored on: {tensor.device}")
```

## Tesnor Operations

Over 100 tensor operations, including transposing, indexing, slicing, mathematical operations, linear algebra, random sampling, and more are comprehensively described [here](https://pytorch.org/docs/stable/torch.html).
<br><br>
Each of them can be run on the GPU (at typically higher speeds than on a CPU). If youâ€™re using Colab, allocate a GPU by going to Edit > Notebook Settings.

```python
import torch

tensor = torch.rand(3, 4)

# We move our tensor to GPU if available
if torch.cuda.is_available():
  tensor = tensor.to('cuda')
  print(f"Device tensor is stored on: {tensor.device}")
```

### **Standard numpy-like indexing and slicing**

```python
tensor = torch.ones(4, 4)
tensor[:, 1] = 0
tensor
```

### **Joining tensors**

You can use `torch.cat` to concatenate a sequence of tensors along a given dimension. See also [`torch.stack`](https://pytorch.org/docs/stable/generated/torch.stack.html), another tensor joing operation that is subtly different from `torch.cat`

```python
t1 = torch.cat([tensor, tensor, tensor], dim=1)
print(t1)
print(f"Shape: {tensor.shape}")
print(f"Shape: {t1.shape}")
```

```python
t2 = torch.cat([tensor, tensor, tensor], dim=0)
print(t2)
print(f"Shape: {tensor.shape}")
print(f"Shape: {t2.shape}")
```

```python
t3 = torch.stack([tensor, tensor, tensor], dim=1)
print(t3)
print(f"Shape: {tensor.shape}")
print(f"Shape: {t3.shape}")
```

```python
t4 = torch.stack([tensor, tensor, tensor], dim=0)
print(t4)
print(f"Shape: {tensor.shape}")
print(f"Shape: {t4.shape}")
```

### **Multiplying tensors**

Element-wise product

```python
# This computs the element-wise product
print(f"tensor.mul(tensor): \n {tensor.mul(tensor)} \n ")

# Alternative syntax:
print(f"tensor * tensor: \n {tensor * tensor} \n")
```

Matrix multiplication

```python
print(f"tensor.matmul(tensor.T): \n {tensor.matmul(tensor.T)} \n")

# Alternative syntax
print(f"tensor @ tensor.T: \n {tensor @ tensor.T} \n")
```

## In-place operations

Operations that have a `_` suffix are in-place.

For example: `x.copy_(y)`, `x.t_()`, will change `x`

```python
print(tensor, "\n")
tensor.add_(5)
print(tensor)
```

> Note: In-place operations save some memory, but can be problematic when computing derivatives beacause of an immediate loss of history, Hence, their use is discouraged.

## Bridge with NumPy

Tensors on the CPU and NumPY arrays can share their underlying memory locations, and changing one will change the other.

### **Tensor to NumPy array**

```python
import torch
import numpy as np
```

```python
torch.cuda.is_available()
```

```python
t = torch.ones(5)
print(f"t: {t}")
n = t.numpy()
print(f"n: {n}")
```

A Change in tensor reflects in the NumPy array

```python
t.add_(1)
print(f"t: {t}")
print(f"n: {n}")
```

### **NumPy array to Tensor**

```python
n = np.ones(5)
t = torch.from_numpy(n)
print(f"n: {n}")
print(f"t: {t}")
```

Changes in the NumPy array reflects in the tensor

```python
np.add(n, 1, out=n)
print(f"n: {n}")
print(f"t: {t}")
```
